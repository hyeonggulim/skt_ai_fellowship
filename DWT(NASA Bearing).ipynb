# Wavelet 종류
+ 종류에 따른 차이?

```python
#dwt 함수 종류
import pywt
pywt.families()
```

```python
df1['Max'].plot(figsize = (12,6))
plt.title("Raw Data")
```

```python
for wave_type in pywt.families():
    try:
        cA, cD = pywt.dwt(df1['Max'], wave_type+'5')
        pd.Series(cA).plot(figsize = (12,6))#lowpass
        pd.Series(cD).plot(figsize = (12,6)) #highpass
        plt.xlabel("Date-Time")
        plt.title(wave_type,fontsize=20)
        plt.legend(['cA(lowpass)', 'cD(highpass)'],fontsize=15)
        plt.show()
    except:
        continue
```

Wavelet Transform

```python
wave_type = 'db3'
feature = 'Max'
cA1, cD1 = pywt.dwt(df1[feature],wave_type)
cA2, cD2 = pywt.dwt(df2[feature],wave_type)
cA3, cD3 = pywt.dwt(df3[feature],wave_type)
cA4, cD4 = pywt.dwt(df4[feature],wave_type)
col = ['cA1','cD1','cA2','cD2','cA3','cD3','cA4','cD4']
```

```python
merged = pd.concat([pd.Series(cA1),pd.Series(cD1),pd.Series(cA2),pd.Series(cD2),pd.Series(cA3),pd.Series(cD3),pd.Series(cA4),pd.Series(cD4)],axis = 1)
merged.columns = col
merged.head(10)
```

```python
merged.to_csv("nasa_2_max_db3.csv",index= False)
```
